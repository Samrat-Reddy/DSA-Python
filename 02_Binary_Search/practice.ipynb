```python
#01 First Occurrence
import array as myarray
arr = myarray.array('i',[1, 1, 1, 1, 2])

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = low + (high - low) // 2

        if key == arr[mid]:
            result = mid
            high = mid - 1
        elif key >= arr[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return result

print(binary_search(arr, 1))
```

```python
#03 Count Occurrences
import array as myarray
arr = myarray.array('i',[1, 1, 1, 1, 2])

def binary_search(arr, key, find_first):
    low = 0
    high = len(arr) - 1
    bound_index = -1

    while low <= high:
        mid = low + (high - low) // 2

        if key == arr[mid]:
            bound_index = mid
            if find_first:
                high = mid - 1
            else:
                low = mid + 1
        elif key < arr[mid]:
            high = mid - 1
        else:
            low = mid + 1

    return bound_index

def count_occurence(arr, key):
    first_idx = binary_search(arr, key, True)

    if first_idx == -1:
        return 0

    last_idx = binary_search(arr, key, False)
    return (last_idx - first_idx) + 1

print(count_occurence(arr, 1))

```

```python
#04 Lower Bound
import array as myarray
arr = myarray.array('i',[1, 1, 1, 1, 2])

def lower_bound(arr, key):
    low = 0
    high = len(arr) - 1
    ans = len(arr)

    while low <= high:
        mid = low + (high - low) // 2
        if arr[mid] >= key:
            ans = mid
            high = mid - 1
        else:
            low = mid + 1
    return ans

lower_bound(arr, 1)
```
