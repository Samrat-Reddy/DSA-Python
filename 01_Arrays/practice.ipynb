```python
import array as myarray
```

```python
#01 Maximum Element
arr1 = myarray.array('i',[3,7,2,10,4])
max = arr1[0]

for i in range(len(arr1)):
    if arr1[i] > max:
        max = arr1[i]
max
```

```python
#02 Minimum Element
arr2 = myarray.array('i',[3,7,2,10,4])
min = arr2[0]

for i in range(len(arr2)):
    if arr2[i] < min:
        min = arr2[i]
print(min)
```

```python
#03 Count Even Numbers
arr3 = myarray.array('i',[3,7,2,10,4])

count = 0
for i in range(len(arr3)):
    if arr3[i] % 2 == 0:
        count += 1
print(count)
```

```python
#04 Reversing an Array Without Using ::-1
arr4 = myarray.array('i',[3,7,2,10,4])
i = arr4[-1]
while i >= 0: # Wrote while i >= len(arr)
    print(arr4[i])
    i -= 1
```
